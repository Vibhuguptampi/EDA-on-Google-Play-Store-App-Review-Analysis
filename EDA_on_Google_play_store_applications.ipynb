````python
def get_info(df):
  # Creating an empty DataFrame 'func_df' to store information about the columns in the input DataFrame.
  func_df = pd.DataFrame(index=df.columns, columns=["datatype", "total_values", "na_values", "non_na_values", "%na_values", "unique_count"])

  # Retrieving the datatype of each column and assign it to the 'datatype' column in 'func_df'.
  func_df["datatype"] = [type(i) for i in list(df.columns)]

  # Calculating the total number of values in each column and assign it to the 'total_values' column in 'func_df'.
  func_df["total_values"] = [len(df[i]) for i in list(df.columns)]

  # Calculating the number of missing values (NaN) in each column and assign it to the 'na_values' column in 'func_df'.
  func_df["na_values"] = [df[i].isnull().sum() for i in list(df.columns)]

  # Calculating the number of non-missing values in each column and assign it to the 'non_na_values' column in 'func_df'.
  func_df["non_na_values"] = [(~df[i].isnull()).sum() for i in list(df.columns)]

  # Calculating the percentage of missing values in each column and assign it to the '%na_values' column in 'func_df'.
  func_df["%na_values"] = [df[i].isnull().mean() * 100 for i in list(df.columns)]

  # Calculating the number of unique values in each column and assign it to the 'unique_count' column in 'func_df'.
  func_df["unique_count"] = [len(df[i].unique()) for i in list(df.columns)]

  # Returning the resulting DataFrame 'func_df' containing the collected information.
  return func_df
````

Lets deal with some special characters (, $ +) and letters (M,k) present in the Installs, Size, and Price columns

*   all the columns except rating have the object data type but some of the columns like, reviews, size, installs and price have the numerical value. So, we have to transform them into a proper data type and
*   also remove the unwanted values from the numerical columns like ‘+’ and ‘,’ from installs and ‘$’ from price.
*   In the size column we have some values in KB and some values in MB, so we transform all the values in MB.
*  We have string format in Last Updated column. We need to typecaste it into datetime format to deal with it.


````python
# List of character needs to be remove
list_of_chars = ['+', ',' , '$' ]

# List of column names to clean
list_of_columns = ['Installs', 'Price']

# Loop for each column
for col in list_of_columns:

    # Replace each character with an empty string
    for char in list_of_chars:
        store_df[col] = store_df[col].astype(str).str.replace(char, '')

    # Convert col to numeric
    store_df[col] = pd.to_numeric(store_df[col])

# Typecasting the str type to timestamp in "Latest Updated" column.
store_df["Last Updated"]= pd.to_datetime(store_df["Last Updated"])
````

````python
# Display summary information about the 'store_df' DataFrame using the 'info()' method.
store_df.info()
````

Now as you can see in the size column 'M' stands for MB and 'k' stands for KB. it is necessary to convert all the values to a single unit. In this case, we will convert all the units to MB. (1MB = 1024KB)

````python
# defining function to convert all unti in MB and removing unit symbol
def convert(i):
  if 'k' in i:
    return float(i[:-1])/1024
  elif 'M' in i:
    return float(i[:-1])
  else:
    return

store_df['Size']=store_df['Size'].apply(convert)
````
